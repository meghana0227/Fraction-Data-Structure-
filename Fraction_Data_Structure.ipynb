```python
import math
```

```python
class Fraction:

  def __init__(self, num, den):
    if den == 0:
      self.num = None
      self.den = None
    else:
      if (num < 0 and den < 0) or (num > 0 and den > 0):
        self.num= abs(num)
        self.den = abs(den)
      else:
        self.num = -abs(num)
        self.den = abs(den)

  def __valid_fraction(self):
    if self.num == None and self.den == None:
      return False
    else:
      return True

  #string magic methods

  def __str__(self):
    if self.__valid_fraction():
      return f"{self.num}/{self.den}"

    raise ValueError('Invalid Fraction(Denominator is zero)')


  #arithmetic opeators

  def __add__(self, other):
    if self.__valid_fraction() and other.__valid_fraction():
      num = self.num*other.den + other.num*self.den
      den = self.den * other.den

      return Fraction(num, den)

    raise ValueError('Invalid Fraction(Denominator is zero)')

  def __sub__(self, other):
    if self.__valid_fraction() and other.__valid_fraction():
      num = self.num*other.den - other.num*self.den
      den = self.den * other.den

      return Fraction(num, den)

    raise ValueError('Invalid Fraction(Denominator is zero)')

  def __mul__(self, other):
    if self.__valid_fraction() and other.__valid_fraction():
      num = self.num * other.num
      den = self.den * other.den

      return Fraction(num, den)

    raise ValueError('Invalid Fraction(Denominator is zero)')

  def __truediv__(self, other):
    if self.__valid_fraction() and other.__valid_fraction():
      num = self.num * other.den
      den = self.den * other.num

      return Fraction(num, den)

    raise ValueError('Invalid Fraction(Denominator is zero)')

  def __pow__(self, n):
    if self.__valid_fraction():
      num = self.num ** n
      den = self.den ** n

      return Fraction(num, den)

    raise ValueError('Invalid Fraction(Denominator is zero)')

  #comparision magic methods

  def __eq__(self, other):
    if self.__valid_fraction() and other.__valid_fraction():
      if self.num==other.num and self.den==other.den:
        return True
      else:
        return False

    raise ValueError('Invalid Fraction(Denominator is zero)')

  def __ne__(self, other):
    if self.__valid_fraction() and other.__valid_fraction():
      return not self.__eq__(other)

    raise ValueError('Invalid Fraction(Denominator is zero)')

  def __lt__(self, other):
    if self.__valid_fraction() and other.__valid_fraction():
      return self.num/self.den < other.num/other.den

    raise ValueError('Invalid Fraction(Denominator is zero)')

  def __gt__(self, other):
    if self.__valid_fraction() and other.__valid_fraction():
      return self.num/self.den > other.num/other.den

    raise ValueError('Invalid Fraction(Denominator is zero)')

  def __le__(self, other):
    if self.__valid_fraction() and other.__valid_fraction():
      return self.num/self.den <= other.num/other.den

    raise ValueError('Invalid Fraction(Denominator is zero)')

  def __ge__(self, other):
    if self.__valid_fraction() and other.__valid_fraction():
      return self.num/self.den >= other.num/other.den

    raise ValueError('Invalid Fraction(Denominator is zero)')

  #simplification
  def simplify(self):
    if self.__valid_fraction():
      gcd = math.gcd(self.num, self.den)
      self.num = self.num // gcd
      self.den = self.den // gcd
    else:
      raise ValueError('Invalid Fraction(Denominator is zero)')

  #conversion to decimal
  def to_decimal(self):
    if self.__valid_fraction():
      return self.num/self.den
    else:
      raise ValueError('Invalid Fraction(Denominator is zero)')

  #absolute fraction
  def __abs__(self):
    if self.__valid_fraction():
      return Fraction(abs(self.num), abs(self.den))
    else:
      raise ValueError('Invalid Fraction(Denominator is zero)')

  #negation
  def __neg__(self):
    if self.__valid_fraction():
      return Fraction(-self.num, self.den)
    else:
      raise ValueError('Invalid Fraction(Denominator is zero)')

  #mixed fraction
  def mixed_fraction(self):
    if self.__valid_fraction():
      quotient = self.num // self.den
      rem = self.num % self.den

      return f"{quotient} {rem}/{self.den}"
    else:
      raise ValueError('Invalid Fraction(Denominator is zero)')

  #inverse fraction
  def inverse(self):
    if self.__valid_fraction():
      if self.num == 0:
        raise ValueError("Invalid Inverse Fraction(Numerator is zero)")
      elif self.num > 0:
        return Fraction(self.den, self.num)
      else:
        return Fraction(-self.den, abs(self.num))
    else:
      raise ValueError('Invalid Fraction(Denominator is zero)')

  #precentage
  def to_percent(self):
    return self.to_decimal()*100
```
